```python
# Importing libaries, picture, and showing picture
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

image = mpimg.imread('../images/exit-ramp.jpg')
plt.imshow(image)
plt.show()
```

```python
# Importing OpenCV libraries and converting to grayscale
import cv2
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
plt.show()
```

```python
# Applying Gaussian smoothing/blurring with kernel size 
# cv2.Canny() has a 5x5 Gaussian internally, so we are going with a 3x3 here. Must be an odd number
kernel_size = 3
blur_gray = cv2.GaussianBlur(gray,(kernel_size, kernel_size), 0)

# Displaying blur
plt.imshow(blur_gray, cmap='gray')
plt.show()
```

```python
# Defining parameters for the Canny function and running it
low_threshold = 90
high_threshold = 180
edges = cv2.Canny(blur_gray, low_threshold, high_threshold)

# Displaying the image
plt.imshow(edges, cmap='Greys_r')
plt.show()
```
